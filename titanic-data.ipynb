```python
import pandas as pd

# Example DataFrame
data = {'A': [1, 2, 3, 4], 'B': [5, 6, 7, 8]}
df = pd.DataFrame(data)

# New rows to add
new_data = {'A': [5, 6], 'B': [9, 10]}
new_df = pd.DataFrame(new_data)

# Add new rows to the DataFrame using pd.concat()
df = pd.concat([df, new_df], ignore_index=True)

print(df)
```

```python
import pandas as pd

# Example DataFrame with an 'Age' column
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Age': [25, 30, 35, 40, 45],
        'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
```

```python
# Change the 'Age' for the second row (index 1) to 32
df.iloc[1, 1] = 32

print("\nUpdated DataFrame after changing Bob's Age:")
print(df)
```
